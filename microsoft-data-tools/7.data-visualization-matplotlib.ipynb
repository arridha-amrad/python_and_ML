# Checking for correlation with visualization
- Scatterplots
    * common char type
    * a dot is drawn for each intersection between x and y

```python
# we need this module to draw a chart
import matplotlib.pyplot as plt

import pandas as pd
```

```python
delays_df = pd.read_csv("Data/Lots_of_flight_data.csv")
print(delays_df.head())
```

# Check if there is a realtionship between distance and arrival_delay


```python
delays_df.plot(
    kind="scatter", 
    x="DISTANCE", 
    y="ARR_DELAY", 
    color="blue", 
    alpha=0.3, 
    title="Correlation between arrival and distance"
)
plt.show()
```

# From the chart above we could say there is no correlation between distance and arrival_delay

# Now, let's see the correlation between departure delay with arrival_delay

```python
delays_df.plot(
    kind="scatter", 
    x="DEP_DELAY", 
    y="ARR_DELAY", 
    color="green", 
    alpha=0.3, 
    title="Correlation between arrival and distance"
)
plt.show()
```
